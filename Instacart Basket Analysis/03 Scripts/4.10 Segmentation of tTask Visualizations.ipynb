02 Visualize Order Frequency for Key Customer Segments

```python
# Count the number of orders by frequency and parental status
freq_parent = df.groupby(['frequency_flag', 'Parental status']).size().unstack(fill_value=0)
```

```python
# Plotting
ax = freq_parent.plot(kind='bar', stacked=False, figsize=(10, 7), colormap='viridis')
```

```python
#create path
path = r'/Users/Owner/Documents/CareerFoundry/Data Immersion/Achievement 4/01-07-2024 Instacart Basket Analysis'
```

```python
#export to file
ax.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'freq_parents.png'),bbox_inches = 'tight')
```
