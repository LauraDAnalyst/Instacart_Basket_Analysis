02 Find the mean number of orders per department

```python

df_ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']})
```

Analysis: The mean number of orders per department is similar in the full dataframe and the subset. This suggests that the subset gives a relatively accurate picture of the average.

03 Create a loyalty flag

```python
#create max order column
df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
pd.options.display.max_rows = None
```

```python
#check output selected columns, left cell output for review purposes
df_ords_prods_merge[['user_id', 'order_number', 'max_order']].head(100)
```

```python
#Create flag "loyal customer"
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#check output
df_ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
#check output selected columns
df_ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

04 Find basic statistics for each loyalty status

```python
#identify price average, min, and max for each loyalty group
df_ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

Interesting. Loyal customers have the lowest product prices and new customers buy higher. As expected, min and max are the same due to limitations programmed into the system.

05 Create spending flags for each customer

```python
#create spending level column
df_ords_prods_merge['spending_level'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#check output selected columns
df_ords_prods_merge[['user_id', 'prices', 'spending_level']].head(60)
```

```python
#rename columns
df_ords_prods_merge.rename(columns={'spending_level':'mean_spend'}, inplace=True)
```

```python
#check output
#check output selected columns
df_ords_prods_merge[['user_id', 'prices', 'mean_spend']].head()
```

```python
#Create flags for low and high spenders
df_ords_prods_merge.loc[df_ords_prods_merge['mean_spend'] < 10, 'spending_flag'] = 'Low spender'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['mean_spend'] >= 10, 'spending_flag'] = 'High spender'
```

```python
#check output
df_ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
#check output selected columns
df_ords_prods_merge[['user_id', 'mean_spend', 'spending_flag']].head(50)
```

The first 50 users are all low spenders. That makes sense given that there are more low spenders than high spenders

06 Identify frequent vs non-frequent customers

```python
#create column for purchase frequency
df_ords_prods_merge['buy_freq'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#check output selected columns
df_ords_prods_merge[['user_id', 'days_since_prior_order', 'buy_freq']].head(60)
```

```python
#Create flags for buy freuqency
df_ords_prods_merge.loc[df_ords_prods_merge['buy_freq'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['buy_freq'] >10) & (df_ords_prods_merge['buy_freq'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
#Create flags for buy freuqency
df_ords_prods_merge.loc[df_ords_prods_merge['buy_freq'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
#check output
df_ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
#check output selected columns
df_ords_prods_merge[['user_id', 'buy_freq', 'frequency_flag']].head(50)
```
